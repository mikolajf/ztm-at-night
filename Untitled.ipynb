```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stops = pd.read_csv('results/stops.txt', encoding = 'cp1250', index_col = 0)
stop_times = pd.read_csv('results/stop_times.txt')
trips = pd.read_csv('results/trips.txt', index_col = 2)
```

```python
print(trips.service_id.unique())
# ['NP' 'NS' 'NO']

trips = trips[trips.service_id.isin(['NP', 'NS', 'NO'])]
print(len(trips))
trips.head()
```

```python
# stop_times = stop_times[stop_times.trip_id.isin(trips.trip_id)]
stop_times = stop_times[stop_times.trip_id.isin(trips.index)]

stop_times['arrival_time_s'] = (stop_times['arrival_time'].str.extract("(\d{2}):\d{2}:\d{2}").astype(int)*60*60 + 
    stop_times['arrival_time'].str.extract("\d{2}:(\d{2}):\d{2}").astype(int)*60 + 
    stop_times['arrival_time'].str.extract("\d{2}:\d{2}:(\d{2})").astype(int))/60

stop_times["arrival_time_d"] = pd.to_datetime(stop_times.arrival_time_s, unit='m')
stop_times.drop(["arrival_time","arrival_time_s","stop_sequence"], axis = 1, inplace = True)
stop_times.head()
```

```python
stop_times_fill = stop_times.set_index(['trip_id', 'arrival_time_d']).join(trips,on="trip_id")
stop_times_fill = stop_times_fill.groupby(level=0).apply(
    lambda x: x[~x.index.duplicated()].sort_index().reset_index(level=0, drop=True).asfreq("1MIN",method='pad'))
# stop_times_fill = stop_times_fill[stop_times_fill.service_id == "NP"] #decide later
stop_times_fill.head()
```

```python
stop_counts = stop_times_fill.groupby(['stop_id','service_id', 'arrival_time_d']).size().to_frame('count')
# stop_counts = stop_counts.join(stops,on="stop_id")
stop_counts.head()
```

```python
stop_counts_fill = pd.DataFrame(index = pd.MultiIndex.from_product([stop_counts.index.get_level_values(0).unique(),
                                                                    stop_counts.index.get_level_values(1).unique(),
                                       pd.date_range(start=stop_times.arrival_time_d.min(),
                                                     end=stop_times.arrival_time_d.max(),
                                                     freq='min')], names=['stop_id', 'service_id', 'arrival_time_d']))

stop_counts_fill = stop_counts_fill.join(stop_counts, on = ['stop_id', 'service_id', 'arrival_time_d']).fillna(0)
stop_counts_fill = stop_counts_fill.join(stops,on="stop_id")
stop_counts_fill.head()
```

```python
data_to_plot = stop_counts_fill.loc[(slice(None),"NO","1970-01-01 23:45:00"),:].reset_index()
data_to_plot = data_to_plot[["stop_lon","stop_lat","count"]].values
```
