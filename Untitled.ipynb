```python
import pandas as pd
```

```python
stops = pd.read_csv('results/stops.txt', encoding = 'cp1250')
stop_times = pd.read_csv('results/stop_times.txt')
trips = pd.read_csv('results/trips.txt')
```

```python
stops.head()
```

```python
stop_times.head()
```

```python
trips.head()
```

```python
print(trips.service_id.unique())
# ['NP' 'NS' 'NO']

trips = trips[trips.service_id.isin(['NP', 'NS', 'NO'])]
print(len(trips))
trips.head()
```

```python
stop_times = stop_times[stop_times.trip_id.isin(trips.trip_id)]

stop_times['arrival_time_s'] = (stop_times['arrival_time'].str.extract("(\d{2}):\d{2}:\d{2}").astype(int)*60*60 + 
    stop_times['arrival_time'].str.extract("\d{2}:(\d{2}):\d{2}").astype(int)*60 + 
    stop_times['arrival_time'].str.extract("\d{2}:\d{2}:(\d{2})").astype(int))/60

stop_times['departure_time_s'] = (stop_times['departure_time'].str.extract("(\d{2}):\d{2}:\d{2}").astype(int)*60*60 + 
    stop_times['departure_time'].str.extract("\d{2}:(\d{2}):\d{2}").astype(int)*60 + 
    stop_times['departure_time'].str.extract("\d{2}:\d{2}:(\d{2})").astype(int))/60

stop_times.head()
```

```python
all(stop_times.arrival_time_s == stop_times.departure_time_s)
# True

stop_times.drop(["departure_time","departure_time_s"], axis = 1, inplace = True)
```

```python
print(min(stop_times.arrival_time_s),max(stop_times.arrival_time_s))
```

```python
stops_times_count = stop_times.groupby(["stop_id","arrival_time_s"]).size()
stops_times_count
```
